```python
import pandas as pd
df=pd.read_csv("Cel.csv")
df
```

```python
X=df[["First_Letter","First_Two_Letter","First_Three_Letter","First_Four_Letter","Last_Letter","Last_Two_Letter","Last_Three_Letter","Last_Four_Letter","Second_Letter","Second_Letter_Type","Third_Letter","Third_Letter_Type","Fourth_Letter","Fourth_Letter_Type","Second_Last_Letter","Second_Last_Letter_Type","Third_Last_Letter","Third_Last_Letter_Type","Fourth_Last_Letter","Fourth_Last_Letter_Type","Length"]]
y=df[["Suffix"]]
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split


label_encoders = {}
for column in X.columns:
    encoder = LabelEncoder()
    X[column] = encoder.fit_transform(X[column])
    label_encoders[column] = encoder

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
X_train=X_train.reset_index()
del X_train["index"]
X_test=X_test.reset_index()
del X_test["index"]
y_train=y_train.reset_index()
del y_train["index"]
y_test=y_test.reset_index()
del y_test["index"]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import CategoricalNB
# Define the parameter grid to search
param_grid = {
    'alpha': [0.19,0.2,0.25,0.4,0.5,1, 2,3,4,5,6],  # You can adjust these values based on your problem
    'min_categories':range(40,70)
}

# Create a Categorical Naive Bayes classifier
classifier = CategoricalNB()

# Create Grid Search CV object with the classifier, parameter grid, and cross-validation folds
grid_search = GridSearchCV(estimator=classifier, param_grid=param_grid, cv=5)  # 5-fold cross-validation

# Perform Grid Search CV on the data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters and corresponding accuracy
```

```python
fd=grid_search.best_estimator_
fd
```

```python

grid_search.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import CategoricalNB
clf = CategoricalNB(alpha=0.2,min_categories=60).fit(X_train, y_train)
clf.score(X_test,y_test)
```

```python
clf = CategoricalNB(alpha=0.2,min_categories=60).fit(X, y)
clf.score(X,y)
```

```python
import pickle

model_pkl_file = "nb_1_model.pkl"  


with open(model_pkl_file, 'wb') as file:  
    pickle.dump(clf, file)
```
